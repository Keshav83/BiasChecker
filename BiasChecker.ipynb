```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Text2text generation pipeline to neutralize biased sentences.
# You can try "t5-base" for potentially better results.
neutralizer = pipeline("text2text-generation", model="t5-small")

candidate_labels = ["biased", "non-biased"]

def detect_bias(text):
    result = classifier(text, candidate_labels)
    return result

def correct_bias(text):
    # Improved prompt that explicitly asks for respectful, neutral language.
    prompt = f"Rephrase the following sentence to remove bias and be respectful and neutral: '{text}'"
    result = neutralizer(prompt, max_length=200, do_sample=False)
    return result[0]['generated_text']

def bias_chatbot():
    print("BiasChecker Chatbot: Hi! I'm BiasChecker. Tell me something, and I'll check if it sounds biased.")
    print("If it is, I'll also provide a more neutral version. Type 'quit' or 'exit' to end our chat.\n")
    while True:
        user_input = input("You: ")
        if user_input.lower() in ['quit', 'exit']:
            print("BiasChecker Chatbot: Goodbye!")
            break

        result = detect_bias(user_input)
        predicted_label = result['labels'][0]
        confidence = result['scores'][0]

        if predicted_label == "biased":
            corrected_text = correct_bias(user_input)
            print(f"BiasChecker Chatbot: I think your message is {predicted_label} (confidence: {confidence:.2f}).")
            print("Here's a more neutral version:")
            print(corrected_text + "\n")
        else:
            print(f"BiasChecker Chatbot: I think your message is {predicted_label} (confidence: {confidence:.2f}).\n")

bias_chatbot()
```
